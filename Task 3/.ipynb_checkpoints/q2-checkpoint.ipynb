```python
import random
import numpy as np
import matplotlib.pyplot as plt

data_file = open("four_circle.txt", "r")
data = []
for line in data_file:
    curr_line = line.split()
    data.append([float(curr_line[0]), float(curr_line[1]), int(curr_line[2])])
data_file.close()
colors = ["blue", "red"]


def plot_set_with_line(lines):
    for p in range(len(data)):
        index = int(max(0, data[p][2]))
        color = colors[index]
        xp = data[p][0]
        yp = data[p][1]
        plt.scatter(xp, yp, c=color, marker='o', linewidths=2)
    plt.grid()
    for x1, y1, x2, y2, _, _ in lines:
        m, n = line_equation(x1, y1, x2, y2)
        if m is None:
            plt.plot([n, n], [0, 4], color='black', linestyle='-', linewidth=1)
        else:
            x = np.linspace(-4, 4, 100)
            y = m * x + n
            plt.plot(x, y, '-g')
    plt.xlim([0, 4])
    plt.ylim([0, 4])
    plt.show()


def split_to_train_test():
    S = []
    T = []
    for n in data:
        if (random.getrandbits(1) == 1 and len(S) < (len(data) / 2)) or len(T) >= (len(data) / 2):
            S.append(n)
        else:
            T.append(n)

    return S, T


def set_rule(S):
    H = []
    x, y = 0, 1
    for first_point in range(len(S)):
        for second_point in range(first_point + 1, len(S)):
            x1 = S[first_point][x]
            y1 = S[first_point][y]
            x2 = S[second_point][x]
            y2 = S[second_point][y]
            H.append((x1, y1, x2, y2))
    return H


def line_equation(x1, y1, x2, y2):
    if np.abs(x1 - x2) < 0.0001:
        m = None
        n = x1
    else:
        m = (y1 - y2) / (x1 - x2)
        n = y1 - (m * x1)
    return m, n


def get_empirial_error(weight, H):
    eth_pos_array = []
    eth_neg_array = []
    hj_xi = []

    for x1, y1, x2, y2 in H:
        hj_pos = []
        hj_neg = []
        m, n = line_equation(x1, y1, x2, y2)
        eth_pos = 0
        eth_neg = 0
        if m is not None:
            for xi, yi, labeli in S:
                # Point under the line
                h_xi = -1
                if ((m * xi) + n) - yi < 0:
                    # Point above the line
                    h_xi = 1
                if h_xi != labeli:
                    eth_pos += weight[S.index([xi, yi, labeli])]
                else:
                    eth_neg += weight[S.index([xi, yi, labeli])]
                hj_pos.append(h_xi)
                hj_neg.append(h_xi * -1)
        else:
            for xi, yi, labeli in S:
                # Point on the left
                h_xi = -1
                if xi > n:
                    # Point on the right
                    h_xi = 1

                if h_xi != labeli:
                    eth_pos += weight[S.index([xi, yi, labeli])]
                else:
                    eth_neg += weight[S.index([xi, yi, labeli])]
                hj_pos.append(h_xi)
                hj_neg.append(h_xi * -1)

        hj_xi.append([hj_pos, hj_neg])
        eth_pos_array.append(eth_pos)
        eth_neg_array.append(eth_neg)
    return eth_pos_array, eth_neg_array, hj_xi


def adaboost(S):
    # Initial the weights to 1/N
    weight = [1 / len(S)] * len(S)
    H = set_rule(S)
    K_rules = []
    for j in range(rule_size):
        # Calculate the imperial error of all the rules
        eth_pos_array, eth_neg_array, hj_xi = get_empirial_error(weight, H)
        # Calculate the min empirical error
        et_ht_pos = min(eth_pos_array)
        et_ht_neg = min(eth_neg_array)
        # Collect the index of min empirical error
        if et_ht_pos < et_ht_neg:
            flag = 1
            index = 0
            err_t = et_ht_pos
            ht = eth_pos_array.index(et_ht_pos)
        else:
            flag = -1
            index = 1
            err_t = et_ht_neg
            ht = eth_neg_array.index(et_ht_neg)

        # Find the alpha
        alpha = 0.5 * np.log((1 - err_t) / err_t)
        # Add the best error
        K_rules.append([H[ht][0], H[ht][1], H[ht][2], H[ht][3], alpha, flag])
        # Update the weights
        for i in range(len(weight)):
            ht_xi = hj_xi[ht][index][i]
            weight[i] = weight[i] * np.exp(-alpha * ht_xi * S[i][2])
        for i in range(len(weight)):
            weight[i] = weight[i] / sum(weight)

    return K_rules


def predict_hi(xi, yi, m, n):
    # Below the line
    ans = -1
    if m is not None:
        # Above the line
        if ((m * xi) + n) - yi <= 0:
            ans = 1
    else:
        if xi > n:
            ans = 1
    return ans


def get_sign(rules, xi, yi):
    sigma = 0
    x1, y1, x2, y2 = 0, 1, 2, 3
    for i in range(len(rules)):
        m, n = line_equation(rules[i][x1], rules[i][y1], rules[i][x2], rules[i][y2])
        val = predict_hi(xi, yi, m, n)
        alpha, flag = 4, 5
        if rules[i][flag] == -1:
            val = val * (-1)
        sigma += rules[i][alpha] * val
    if sigma > 0:
        return 1
    else:
        return -1


def check_error(set_points, rules):
    errors = []
    for i in range(len(rules)):
        count = 0
        for p in set_points:
            pred = get_sign(rules[i], p[0], p[1])
            if pred != p[2]:
                count += 1
        errors.append(count / len(set_points))
    return errors


if __name__ == "__main__":
    rule_size = 8
    run_size = 10
    empirical_error = np.zeros(rule_size + 1)
    true_error = np.zeros(rule_size + 1)

    for i in range(run_size):
        # Split the data
        S, T = split_to_train_test()
        # Run adaboost on train set
        res = adaboost(S)
        H = []
        # Divide the rules into groups [[1], [1,2], .... , [1,2,3,4,5,6,7,8]]
        for j in range(len(res) + 1):
            H.append(res[0:j])
        # Use this command to plot the chart with line: for i1 in H: plot_set_with_line(i1)
        # Check the empirical error on train
        emp_err_i = check_error(S, H)
        # Check the true error on test
        true_err_i = check_error(T, H)
        # Add values to the main arrays
        for j in range(rule_size + 1):
            empirical_error[j] += emp_err_i[j]
            true_error[j] += true_err_i[j]
        print(f'{i + 1} Adaboost run over.')
    print("==========================================")
    print("================= Result =================")
    print("==========================================")
    print()
    # Calculate the average error
    avg_empirical_error = [element / run_size for element in empirical_error]
    avg_true_error = [element / run_size for element in true_error]
    # Print the 16 result we get
    print('Average of empirical error(Train):')
    print(avg_empirical_error)
    print('Average of true error(Test):')
    print(avg_true_error)
    # Show the result on chart
    plt.title('Train vs Test')
    plt.plot(avg_empirical_error, label='Train')
    plt.plot(avg_true_error, label='Test')
    plt.xlabel("Set Of Rules")
    plt.ylabel("Error")
    plt.legend()
    plt.show()
```
